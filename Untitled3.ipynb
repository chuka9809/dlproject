```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, \
    f1_score, precision_score, recall_score, roc_auc_score, roc_curve
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from category_encoders import TargetEncoder
```

```python

df = pd.read_csv('BankChurners.csv')
print("df.shape: ", df.shape)
df.head()
```

```python
# 의미없는 행 제거
df = df.iloc[:, 1:-2]
df.info()
```

```python

df['Income_Category'].value_counts()
```

```python

#TODO: flag 변수에 대해 추후 미세한 조정 필요

num_cols = df.select_dtypes(include=np.number).columns.to_list()
cat_cols = df.select_dtypes(exclude=np.number).columns.to_list()
print("num_cols: ", num_cols)
print("cat_cols: ", cat_cols)
```

```python

codes = {'Existing Customer':0, 'Attrited Customer':1}
df_clean = df.copy()
df_clean['Attrition_Flag'] = df['Attrition_Flag'].map(codes).astype(int)

extended_cols = num_cols.copy()
extended_cols.insert(0, 'Attrition_Flag')

sns.set_theme(rc={"figure.figsize": (15, 15)})
sns.heatmap(
    df_clean[extended_cols].corr(),
    annot=True,
    linewidths=0.5,
    center=0,
    cbar=False,
    cmap="YlGnBu",
    fmt="0.2f"
)
plt.show()
```

```python

df_model = df.copy()
df_model['Attrition_Flag'] = df_model['Attrition_Flag'].map(codes).astype(int)

# columns_to_drop = ["Credit_Limit", "Dependent_count", "Months_on_book", "Avg_Open_To_Buy", "Customer_Age"]
# df_model = df_model.drop(columns=columns_to_drop)

one_hot_encoding_feature = ['Gender']
ohe = OneHotEncoder(sparse_output=False, drop='first')
df_ohe = pd.DataFrame(ohe.fit_transform(df_model[one_hot_encoding_feature]), columns=ohe.get_feature_names_out(one_hot_encoding_feature))
df_model = df_model.drop(columns=one_hot_encoding_feature).reset_index(drop=True)
df_model = pd.concat([df_model, df_ohe], axis=1)

ordinal_encoding_orders = {
    'Education_Level': ['Uneducated', 'High School', 'College', 'Graduate', 'Post-Graduate', 'Doctorate', 'Unknown'],
    'Income_Category': ['Less than $40K', '$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +', 'Unknown'],
    'Card_Category': ['Blue', 'Silver', 'Gold', 'Platinum']
}

ordinal_features = list(ordinal_encoding_orders.keys())
ordinal_encoder = OrdinalEncoder(categories=[ordinal_encoding_orders[col] for col in ordinal_features], handle_unknown='use_encoded_value', unknown_value=-1)
df_model[ordinal_features] = ordinal_encoder.fit_transform(df_model[ordinal_features])

target_encoding_feature = ['Marital_Status']
target_encoder = TargetEncoder(cols=target_encoding_feature)
df_model[target_encoding_feature] = target_encoder.fit_transform(df_model[target_encoding_feature], df_model['Attrition_Flag'])

df_model.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_model.drop('Attrition_Flag', axis=1),\
       df_model['Attrition_Flag'], test_size=0.2, random_state=42, stratify=df_model['Attrition_Flag'])
```

```python

def get_scores(y_test, y_pred):
    scores = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'ROC AUC': roc_auc_score(y_test, y_pred),
        'F1 Score': f1_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred)
    }
    return scores
```

```python

def print_scores(y_test, y_pred):
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print("Accuracy: ", accuracy_score(y_test, y_pred))
    print("ROC_AUC: ", roc_auc_score(y_test, y_pred))
    print("F1 Score: ", f1_score(y_test, y_pred))
    print("Precision: ", precision_score(y_test, y_pred))
    print("Recall: ", recall_score(y_test, y_pred))
    print("ROC AUC: ", roc_auc_score(y_test, y_pred))
```

```python

def plot_model_comparison(rf_scores, xgb_scores, label1, label2):
    labels = list(rf_scores.keys())
    rf_values = list(rf_scores.values())
    xgb_values = list(xgb_scores.values())

    x = range(len(labels))

    plt.figure(figsize=(10, 6))
    plt.bar(x, rf_values, width=0.4, label=label1, align='center')
    plt.bar(x, xgb_values, width=0.4, label=label2, align='edge')
    plt.xlabel('Metrics')
    plt.ylabel('Scores')
    plt.title(f"{label1} vs {label2}")
    plt.xticks(x, labels, rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python

rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
rf_clf = rf_clf.fit(X_train, y_train)
y_pred = rf_clf.predict(X_test)
print_scores(y_test, y_pred)
rf_scores = get_scores(y_test, y_pred)
```

```python

xgb_clf = XGBClassifier()
xgb_clf.fit(X_train, y_train)
y_pred = xgb_clf.predict(X_test)
print_scores(y_test, y_pred)
xgb_scores = get_scores(y_test, y_pred)
```

```python

plot_model_comparison(rf_scores, xgb_scores, 'Random Forest', 'XGBoost')
```

```python

param_dist = {
    'n_estimators': randint(100, 500),
    'learning_rate': uniform(0.01, 0.2),
    'max_depth': randint(3, 10),
    'subsample': uniform(0.6, 0.4),
    'colsample_bytree': uniform(0.6, 0.4)
}

xgb_clf = XGBClassifier(random_state=42)

random_search = RandomizedSearchCV(estimator=xgb_clf, param_distributions=param_dist, n_iter=100, scoring='recall', cv=5, verbose=1, n_jobs=-1, random_state=42)

random_search.fit(X_train, y_train)
print("Best parameters found: ", random_search.best_params_)

best_xgb_clf = random_search.best_estimator_
y_pred = best_xgb_clf.predict(X_test)
print_scores(y_test, y_pred)
best_xgb_scores = get_scores(y_test, y_pred)
```

```python

plot_model_comparison(xgb_scores, best_xgb_scores, 'xbg_scores', 'best_xgb_scores')
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.datasets import make_classification

print('Resampled dataset shape %s' % Counter(y_train))
# SMOTE 적용
sm = SMOTE(random_state=42)
X_train, y_train = sm.fit_resample(X_train, y_train)

print('Resampled dataset shape %s' % Counter(y_train))
```

```python
xgb_clf = XGBClassifier()
xgb_clf.fit(X_train, y_train)
y_pred = xgb_clf.predict(X_test)
print_scores(y_test, y_pred)
xgb_smote_scores = get_scores(y_test, y_pred)
```

```python
plot_model_comparison(xgb_scores, xgb_smote_scores, 'XGBoost', 'XGBoost after smote')
plot_model_comparison(best_xgb_scores, xgb_smote_scores, 'Best XGBoost', 'XGBoost after smote')

```
